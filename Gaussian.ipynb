```python
import numpy as np
import cv2
from PIL import *
from matplotlib import pyplot as plt
import math
from scipy import signal # for convolution
```

## Gaussian Mask and Derivative

```python
def g(X, Y, sigma):
    return np.exp(-(X**2 + Y**2)/(2*sigma**2))

def gx(X, Y, sigma):
    return np.exp(-(X**2 + Y**2)/(2*sigma**2))*(-X/(sigma**2))

def gy(X, Y, sigma):
    return np.exp(-(X**2 + Y**2)/(2*sigma**2))*(-Y/(sigma**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = g(X, Y,3)
Zx = gx(X,Y,3)
Zy = gy(X,Y,3)
```

```python
def draw_save_fig(X, Y, Z, label_z, view_i, view_j, name, title):
    fig = plt.figure()
    fig.set_size_inches(18.5, 10.5)
    ax = plt.axes(projection='3d')
    ax.set_title(title)
    ax.contour3D(X, Y, Z, 200)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel(label_z, labelpad=20);
    ax.view_init(view_i, view_j)
    fig.savefig(name+".png")
```

```python
draw_save_fig(X, Y, Z, "z", 0, 0, "G", "G")
draw_save_fig(X, Y, Zx, "zx", 0, 100, "Gx","Gx")
draw_save_fig(X, Y, Zy, "zy", 0, 0, "Gy","Gy")
```

```python
def gx2(X, Y, sigma):
    return np.exp(-(X**2 + Y**2)/(2*sigma**2))*(1/sigma**2)*(-1+X**2/(sigma**2))

def gy2(X, Y, sigma):
    return np.exp(-(X**2 + Y**2)/(2*sigma**2))*(1/sigma**2)*(-1+Y**2/(sigma**2))

Zx2 = gx2(X, Y, 3)
Zy2 = gy2(X, Y, 3)
```

```python
draw_save_fig(X,Y,Zx2, "zx2", 0, 100, "Gx2", "Gx2")
draw_save_fig(X,Y,Zy2, "zy2", 0, 0, "Gy2","Gy2")
```
